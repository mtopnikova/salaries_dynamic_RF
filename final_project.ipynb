## Описание проекта
Данные по среднемесячной ЗП работников организаций по видам экономической деятельности \
в Российской Федерации за 2000-2023 гг были взяты с сайта Росстата: \
(https://rosstat.gov.ru/labor_market_employment_salaries) \
Данные об уровне инфляции были взяты с сайта: \
(https://xn----ctbjnaatncev9av3a8f8b.xn--p1ai/%D1%82%D0%B0%D0%B1%D0%BB%D0%B8%D1%86%D1%8B-%D0%B8%D0%BD%D1%84%D0%BB%D1%8F%D1%86%D0%B8%D0%B8) \
Из всех видов деятельности только для 11 видов деятельности в таблицах была приведена корректная \
информация по зарплатам за весь промежуток 2000-2023 гг. Из них были выбраны три вида деятельности \
(образование, здравоохранение и социальные услуги и добыча полезных ископаемых). \
Для этих трех видов деятельности было проведено сравнение динамики номинальных зарплат с динамикой \
реальных зарплат (реальные зарплаты рассчитывались с учетом уровня инфляции с использованием \
сложного процента, то есть все зарплаты пересчитывались на уровень цен 2000 года) 


```python
import pandas as pd
import plotly.express as px
```

```python
df1 = pd.read_excel('./RF_salaries.xlsx', skiprows=range(4))
df1 = df1.rename({'Unnamed: 0': 'Вид деятельности'}, axis=1)
df1.dropna(inplace=True)
df1.head()
```

```python
df2 = pd.read_excel('RF_salaries.xlsx', sheet_name=1, skiprows=range(2))
df2.rename({'Unnamed: 0': 'Вид деятельности'}, axis=1, inplace=True)
df2.dropna(inplace=True)
df2.head()
```

```python
df = pd.merge(left=df2, right=df1, on='Вид деятельности')
df
```

```python
areas = [
    'Образование', 'Здравоохранение и предоставление социальных услуг',
    'Добыча полезных ископаемых'
]
my_df = df.loc[(df['Вид деятельности'] == areas[0]) |
               (df['Вид деятельности'] == areas[1]) |
               (df['Вид деятельности'] == areas[2])]
my_df = my_df.reset_index(drop=True).rename({'Вид деятельности': 'Год'},
                                            axis=1)
my_df = my_df.set_index('Год').transpose()
```

```python
my_df
```

```python
my_df['Годовой уровень инфляции'] = [
    0, 18.58, 15.06, 11.99, 11.74, 10.91, 9.00, 11.87, 13.28, 8.80, 8.78, 6.10,
    6.58, 6.45, 11.36, 12.91, 5.38, 2.52, 4.27, 3.05, 4.91, 8.39, 11.92, 7.42
]
```

```python
my_df['Coef'] = 1 - my_df['Годовой уровень инфляции'] / 100
```

```python
for area in areas:
    my_df[f'{area} (с учетом инфляции)'] = my_df[f'{area}'] * my_df[
        'Coef'].cumprod()
    my_df[f'{area} (с учетом инфляции)'] = my_df[
        f'{area} (с учетом инфляции)'].round(2)
```

```python
my_df
```

# Графики
Построено 5 графиков, отражающих:
- динамику номинальных зарплат для всех 3 видов деятельности
- динамику реальных зарплат для всех 3 видов деятельности
- сравнение динамики номинальной и реальной ЗП в сфере образования
- сравнение динамики номинальной и реальной ЗП в сфере добычи полезных ископаемых
- сравнение динамики номинальной и реальной ЗП в сфере здравоохранения и социальных услуг 

```python
# график для номинальных зарплат

fig1 = px.line(my_df,
               x=my_df.index,
               y=my_df.columns[:3],
               labels={
                   'index': 'Год',
                   'value': 'Средняя зарплата',
                   'variable': 'Вид деятельности'
               },
               markers=True,
               height=650,
               width=1000)
fig1.update_layout(
    title=
    'Динамика номинальной среднемесячной ЗП работников в РФ за 2000-2023 гг',
    title_font_size=17,
    legend=dict(orientation="v",
                yanchor="top",
                y=0.95,
                xanchor="left",
                x=0.05,
                title_text=''))
fig1.show()
```

```python
# график ЗП с учетом инфляции
fig2 = px.line(my_df,
               x=my_df.index,
               y=my_df.columns[5:],
               labels={
                   'index': 'Год',
                   'value': 'Средняя зарплата',
                   'variable': 'Вид деятельности'
               },
               markers=True,
               height=650,
               width=1000)
fig2.update_layout(
    title='Динамика реальной среднемесячной ЗП работников в РФ за 2000-2023 гг',
    title_font_size=17,
    legend=dict(orientation="v",
                yanchor="top",
                y=0.98,
                xanchor="left",
                x=0.01,
                title_text=''))
fig2.show()
```

```python
# образование

fig3 = px.line(my_df,
               x=my_df.index,
               y=my_df.columns[[1, 5]],
               labels={
                   'index': 'Год',
                   'value': 'Средняя зарплата',
                   'variable': 'Вид деятельности'
               },
               markers=True)
fig3.update_layout(
    title=
    'Динамика среднемесячной ЗП работников в сфере образования в РФ за 2000-2023 гг',
    title_font_size=14,
    legend=dict(orientation="v",
                yanchor="top",
                y=0.95,
                xanchor="left",
                x=0.05,
                title_text=''))

newnames = {
    'Образование': 'Номинальная ЗП',
    'Образование (с учетом инфляции)': 'Реальная ЗП (с учетом инфляции)'
}
fig3.for_each_trace(lambda t: t.update(name=newnames[t.name],
                                       legendgroup=newnames[t.name],
                                       hovertemplate=t.hovertemplate.replace(
                                           t.name, newnames[t.name])))
fig3.show()
```

```python
# добыча полезных ископаемых

fig4 = px.line(my_df,
               x=my_df.index,
               y=my_df.columns[[0, 7]],
               labels={
                   'index': 'Год',
                   'value': 'Средняя зарплата',
                   'variable': 'Вид деятельности'
               },
               markers=True)
fig4.update_layout(
    title=
    'Динамика среднемесячной ЗП работников в сфере добычи полезных ископаемых в РФ за 2000-2023 гг',
    title_font_size=14,
    legend=dict(orientation="v",
                yanchor="top",
                y=0.95,
                xanchor="left",
                x=0.05,
                title_text=''))

newnames = {
    'Добыча полезных ископаемых':
    'Номинальная ЗП',
    'Добыча полезных ископаемых (с учетом инфляции)':
    'Реальная ЗП (с учетом инфляции)'
}
fig4.for_each_trace(lambda t: t.update(name=newnames[t.name],
                                       legendgroup=newnames[t.name],
                                       hovertemplate=t.hovertemplate.replace(
                                           t.name, newnames[t.name])))
fig4.show()
```

```python
# здравоохранение и социальные услуги

fig5 = px.line(my_df,
               x=my_df.index,
               y=my_df.columns[[2, 6]],
               labels={
                   'index': 'Год',
                   'value': 'Средняя зарплата',
                   'variable': 'Вид деятельности'
               },
               markers=True)
fig5.update_layout(
    title=
    'Динамика среднемесячной ЗП работников в сфере здравоохранения и социальных услуг в РФ за 2000-2023 гг',
    title_font_size=14,
    legend=dict(orientation="v",
                yanchor="top",
                y=0.95,
                xanchor="left",
                x=0.05,
                title_text=''))

newnames = {
    'Здравоохранение и предоставление социальных услуг':
    'Номинальная ЗП',
    'Здравоохранение и предоставление социальных услуг (с учетом инфляции)':
    'Реальная ЗП (с учетом инфляции)'
}
fig5.for_each_trace(lambda t: t.update(name=newnames[t.name],
                                       legendgroup=newnames[t.name],
                                       hovertemplate=t.hovertemplate.replace(
                                           t.name, newnames[t.name])))
fig5.show()
```

Номинальные среднемесячные ЗП работников всех трех видов деятельности стабильно росли с 2000 по 2023 год. \
